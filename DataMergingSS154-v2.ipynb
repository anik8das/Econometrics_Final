```python
# pip install pandas 
import pandas as pd
import matplotlib.pyplot as plt

# Read the csv file
data = pd.read_csv('MinimumWageData.csv', encoding= 'unicode_escape')
data["Year"] = data[["Year"]].astype(int)
data = data[data["Year"]>=1976]
```

```python
data.head()
```

```python
data = data.rename(columns={\
        "State.Minimum.Wage" : "state_minwage", "State.Minimum.Wage.2020.Dollars" : "state_minwage_2020d", "Federal.Minimum.Wage":"fed_minwage",\
            "Federal.Minimum.Wage.2020.Dollars":"fed_minwage_2020d", "Effective.Minimum.Wage":"eff_minwage", "Effective.Minimum.Wage.2020.Dollars":"eff_minwage_2020d",\
                "CPI.Average":"cpi_avg", "Department.Of.Labor.Uncleaned.Data":"uncleaned_dol", "Department.Of.Labor.Cleaned.Low.Value":"dol_low",\
                    "Department.Of.Labor.Cleaned.Low.Value.2020.Dollars":"dol_low_2020d", "Department.Of.Labor.Cleaned.High.Value":"dol_high",\
                        "Department.Of.Labor.Cleaned.High.Value.2020.Dollars":"dol_high_2020d"})
data = data.drop(["Footnote"], axis = 1)
data.head()
```

```python
def state_df(state_name: str, filename : str = None):
    if filename is None:
        filename = state_name.lower()
    df = pd.read_csv(f'{filename.lower()}.csv')
    df['DATE'] = df['DATE'].str[:4]
    df['DATE'] = df[['DATE']].astype(int)
    df = df.groupby("DATE").mean().reset_index()
    df = df.rename(columns={df.columns[1]: 'Unemployment', 'DATE':'Year'})
    df['State'] = state_name
    return df
```

```python
dataNY = state_df("New York", "ny")
dataMass = state_df("Massachusetts", "mass")
dataColorado = state_df("Colorado")
dataPenn = state_df("Pennsylvania", 'penn')
dataWV = state_df("West Virginia", 'vaur')
dataWV = state_df("Florida", 'flur')
dataWV = state_df("Ohio", 'ohur')
dataWV = state_df("Alabama", 'alur')
dataWV = state_df("Wisconsis", 'wiur')
dataWV = state_df("Michigan", 'miur')
dataWV = state_df("Illinois", 'ilur')
dataTexas = state_df("Texas")
dataTexas.head()
```

```python
combinedFrames = [dataColorado, dataMass, dataTexas, dataNY, dataPenn, dataWV]
stateUnemploymentData = pd.concat(combinedFrames)
```

```python
result = pd.merge(data, stateUnemploymentData, on=["State", "Year"])
print(result[0:10])
result.to_csv("final.csv")
```
